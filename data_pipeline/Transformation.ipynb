## Handle Null Value

| Column                              | What to do                  |
| ----------------------------------- | --------------------------- |
| phone, email, address,specialization              | Fill with `"Not Available"` |
| personal_links                        |  Fill with `"Not Provided"`     |
| journal_articles, conference_papers | Fill with `0`   |
| bio, teaching                       | Fill with `"Not Provided"`  |


```python
df.fillna({
    "phone": "Not Available",
    "email": "Not Available",
    "address": "Not Available",
    "specialization": "Not Available",
    "bio": "Not Provided",
    "teaching": "Not Provided",
    "personal_links": "Not Provided",
    "research_areas":  "Not Provided",
    "journal_articles": "0",
    "conference_papers": "0"
}, inplace=True)
```

```python
def fix_email(email):
    email = email.replace("[at]", "@")
    email = email.replace("[dot]", ".")
    return email

df["email"] = df["email"].apply(fix_email)
```

```python
def specialization_to_list(x):

    if pd.isna(x) or x=="Not Provided":
        return x

    x = str(x)

    return [i.strip() for i in x.split(",") if i.strip()!=""]


df["specialization"] = df["specialization"].apply(specialization_to_list)
```

```python
import re

BLACKLIST = [
    "assistant", "assistant professor", "adjunct faculty",
    "teaching assistant", "professor", "role",
    "da-iict", "gandhinagar", "florida tech",
    "niser", "iit", "india", "usa"
]


def clean_teaching_list(x):

    if pd.isna(x):
        return "Not Provided"

    x = str(x)

    # remove brackets & quotes
    x = x.strip("[]").replace("'", "")

    # split into list
    items = [i.strip() for i in x.split(",") if i.strip() != ""]

    cleaned = []

    for item in items:

        item = item.replace("\xa0"," ").strip()
        item_lower = item.lower()

        # remove URLs
        if re.search(r'https?://|www\.', item):
            continue

        # remove roles + institutes
        if any(b in item_lower for b in BLACKLIST):
            continue

        cleaned.append(item)

    if not cleaned:
        return "Not Provided"

    return cleaned


df["teaching"] = df["teaching"].apply(clean_teaching_list)
```

```python
import ast

def fix_list(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return x
    return x

df["journal_articles"] = df["journal_articles"].apply(fix_list)
df["conference_papers"] = df["conference_papers"].apply(fix_list)
```

```python
def research_areas_list_to_sting(x):

    if pd.isna(x) or str(x).strip() == "":
        return "Not Provided"

    if x == "Not Provided":
        return x

    if str(x).startswith("["):
        x = ast.literal_eval(x)

    if type(x) == list:
        return " ".join(x)

    return x
df["research_areas"] = df["research_areas"].apply(research_areas_list_to_sting)
```

```python
df.to_json("../data/raw_data.json", orient="records", indent=2)
print("Saved as JSON successfully!")
```

```python
import json
import re

REPLACEMENTS = {
    r"\\xa0": " ",
    r"\\u2013": "",
    r"\\u201c": "",
    r"\\u201d": "",
    r"\\u2019": "'",

    "\xa0": " ",
    "\u2013": "",
    "\u201c": "",
    "\u201d": "",
    "\u2019": "'"
}

def clean_text(text):
    for k, v in REPLACEMENTS.items():
        text = re.sub(k, v, text)
    return re.sub(r"\s+", " ", text).strip()


def deep_clean(obj):

    if type(obj) == str:
        return clean_text(obj)

    if type(obj) == list:
        return [deep_clean(i) for i in obj]

    if type(obj) == dict:
        return {k: deep_clean(v) for k, v in obj.items()}

    return obj


FILE = "../data/raw_data.json"

with open(FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

data = deep_clean(data)

with open(FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```
